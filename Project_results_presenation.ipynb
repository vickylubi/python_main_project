## Estimated Revenue Calculation

Below I proposed two methods for estimating average income in this analysis.

*Important note:*
Actual booking numbers throughout the year are not available; therefore, these figures are intended solely for comparing neighborhoods with each other, not for providing precise revenue forecasts. Additionally, this project does not include data on real estate purchase prices or maintenance costs, so profit calculations were not performed.

### Method 1
The first method for estimating average income is based on a property's availability and price. This approach uses the number of days a listing is available for booking (availability) and multiplies it by the average nightly rate to estimate potential revenue. The formula can be summarized as:

- Estimated Income = Availability (days)  Average Nightly Price

*Important note:* this method does not account for actual occupancy rates and overestimate revenue.

### Method 2

The second method estimates average income based on the property's price and the number of reviews received per month, multiplied by 12 to annualize the figure. This approach uses the monthly review count as a proxy for the number of bookings, under the assumption that each review represents a completed stay. The formula can be summarized as:

- Estimated Annual Income = Average Nightly Price  Reviews per Month  12


### Combining results of both approches
Since Method 1 produced much higher values than Method 2, I adjusted the scales to make them comparable: results from Method 1 were multiplied by 0.2, and results from Method 2 by 0.8. This proportional scaling allowes to average results of both calculations.
