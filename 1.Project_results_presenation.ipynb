## Introduction

With Tokyo’s tourism industry booming, short-term rentals on platforms like Airbnb have become a lucrative investment. However, not all neighborhoods offer the same potential for high returns. 
This analysis aims to answer:  
**Which Tokyo neighborhoods are best suited for Airbnb investment based on price, occupancy, and guest satisfaction?**

東京の観光業が活発になる中で、Airbnbのような短期賃貸は魅力的な投資先となっています。
しかし、すべてのエリアが高い収益を生むわけではありません。
本分析では、以下の問いに答えることを目的としています：
**「価格・稼働率・レビュー評価に基づき、Airbnb投資に最適な東京のエリアはどこか？」**

**Goal:** Assist potential investors in identifying the most promising neighborhoods in Tokyo for launching new Airbnb listings, based on pricing trends, occupancy proxies, and review patterns in the existing Airbnb dataset.

価格の傾向、稼働率の代替指標、レビューの傾向に基づき、
東京で新しいAirbnbを始めるのに最も有望なエリアを投資家向けに提案することを目的としています。

## Estimated Revenue Calculation / 予想収益計算

Below I proposed two methods for estimating average revenue in this analysis.<br>
以下の二つの方法で、平均収益の推定を行いました。

*Important note:*
Actual booking numbers throughout the year are not available; therefore, these figures are intended solely for comparing neighborhoods with each other, not for providing precise revenue forecasts. Additionally, this project does not include data on real estate purchase prices or maintenance costs, so profit calculations were not performed.<br>
*重要な注意点:*
年間を通じた実際の予約数は利用できないため、これらの数値は正確な収益予測を提供するためではなく、地域間の比較のみを目的としています。さらに、このプロジェクトには不動産購入価格や維持管理費用のデータが含まれていないため、利益計算は行っていません。

### Method 1
The first method for estimating average income is based on a property's availability and price. This approach uses the number of days a listing is available for booking (availability) and multiplies it by the average nightly rate to estimate potential revenue. The formula can be summarized as:

- Estimated Income = Availability (days) × Average Nightly Price

*Important note:* this method does not account for actual occupancy rates and overestimate revenue.<br>

平均収益を推定する最初の方法は、物件の空室状況と価格に基づいています。このアプローチでは、リスティングが予約可能な日数（空室日数）を使用し、それに平均的な1泊の料金を掛け算して、潜在的な収益を推定します。式は以下のようにまとめられます：

- 予想収益 = 空室日数（日） × 平均1泊料金<br>

重要な注意点: この方法では実際の占有率を考慮していないため、収益を過大評価する可能性があります。

### Method 2

The second method estimates average income based on the property's price and the number of reviews received per month, multiplied by 12 to annualize the figure. This approach uses the monthly review count as a proxy for the number of bookings, under the assumption that each review represents a completed stay. The formula can be summarized as:

- Estimated Annual Income = Average Nightly Price × Reviews per Month × 12

二番目の方法では、物件の価格と毎月のレビュー数に基づいて平均収益を推定します。この方法では、レビュー数を予約数の代理として使用し、各レビューが完了した宿泊を示しているという仮定のもと、月間レビュー数を12倍して年間収益を推定します。式は以下のようにまとめられます：

- 推定年間収益 = 平均1泊料金 × 月間レビュー数 × 12

### Combining results of both approaches / 両方の方法の結果を組み合わせる
Since Method 1 produced much higher values than Method 2, I adjusted the scales to make them comparable: results from Method 1 were multiplied by 0.2, and results from Method 2 by 0.8. This proportional scaling allowes to average results of both calculations. <br>
方法1の結果は方法2の結果よりもはるかに高い値を示したため、それらを比較可能にするためにスケールを調整しました。方法1の結果には0.2を掛け、方法2の結果には0.8を掛けました。この比率に基づいたスケーリングにより、両方の計算結果を平均化することができました。

## Limitations & Future Work

- The dataset lacks actual booking and revenue data; occupancy is inferred from review frequency and listing availability.
- Local regulations for short-term rentals vary by ward and can affect profitability-thorough legal due diligence is essential.
- Future analysis should incorporate:
    - Seasonality and event-driven demand (e.g., cherry blossom season, major events)
    - Operational costs (cleaning, management, permits) to enable true ROI calculations
    - More granular guest segmentation and pricing strategies

- このデータセットには実際の予約数や収益のデータが含まれておらず、占有率はレビュー頻度とリスティングの空室日数から推測されています。
- 短期賃貸に関する地域ごとの規制は区によって異なり、収益性に影響を与える可能性があります。そのため、法的な調査（デューデリジェンス）が不可欠です。
- 今後の分析では以下の要素を取り入れるべきです：
    - 季節性やイベントに基づく需要（例：桜の季節、主要イベントなど）
    - 清掃・管理・許可などの運営コストを考慮した実際のROI（投資収益率）の計算
    - より詳細なゲストのセグメンテーションや価格戦略の分析

## Conclusion / 結論
Tokyo offers diverse Airbnb investment opportunities, but returns depend heavily on neighborhood dynamics, regulatory environment, and property type. By focusing on data-driven indicators-price, occupancy proxies, and guest satisfaction-investors can make more informed decisions and maximize their potential returns.<br>
東京には多様なAirbnb投資の機会がありますが、収益は地域の特性、規制環境、物件の種類に大きく左右されます。価格、稼働率の代替指標、ゲスト満足度といったデータに基づいた指標に注目することで、投資家はより的確な判断を下し、収益の最大化を図ることができます。
